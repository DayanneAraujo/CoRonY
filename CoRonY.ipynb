```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('covid19.csv')
df
```

```python
df_ontario = df.loc[df['prname'] == 'British Columbia']
```

```python
df_ontario
```

```python
pd.to_datetime(df_ontario['date'])
df_ontario.set_index('date', inplace=True)
```

```python
df_ontario['numconf'].plot(kind='bar')
plt.xticks(rotation=90)
plt.xlabel("Date")
plt.ylabel("Number of cases")
plt.show()
```
